##### To organize one purpose column for each row with the three description columns
- For the three columns of descriptions: "prime_award_base_transaction_description" is the most specific one and looked at first. 
- If the value is na, the the next most specific one: "product_or_service_code_description" is taken to fill the purpose column.
- If that value is also na, then "naics_description" is used.

```python
df_sum["purpose"] = df_sum.prime_award_base_transaction_description.fillna(df_sum.product_or_service_code_description).fillna(df_sum.naics_description)
```

##### Now pivot table it by recipient name to look at how much which recipient gets which year.

```python
df_rec = pd.pivot_table(df_sum, index=["recipient_parent", "award_first_year"], values=["amount"], aggfunc="sum").reset_index()
```

```python
df_rec.head(30)
```

```python
df_rec.to_csv("recipients-year.csv", index=False)
```

##### Now look at how much each recipient gets in total:

```python
df_rec_totals = df_rec.pivot_table(index="recipient_parent", values="amount", aggfunc="sum").reset_index()
```

```python
df_rec_totals.head()
```

```python
df_rec_totals.to_csv("recipients-totals.csv", index=False)
```

##### Categorize by "product_or_service_code_description" and count the number of contracts and sum the amount spent for each purpose

```python
df_purpose = df_sum.pivot_table(index="product_or_service_code_description",  values= "amount", aggfunc=["sum", "count"]).reset_index()
df_purpose.head(30)
```

```python
df_amount_purpose = df_purpose.sort_values(by=("sum", "amount"), ascending=False)
```
