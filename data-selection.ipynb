```python
import pandas as pd
```

```python
df_purpose = pd.read_csv("purpose-amount.csv")
df_rec_totals = pd.read_csv("recipients-totals.csv")
df_rec_year = pd.read_csv("recipients-year.csv")
```

```python
df_rec_totals.head()
```

```python
# look at who received the most money over all the years:
df_totals_max = df_rec_totals.sort_values(by="amount", ascending=False).head(10)
df_totals_max
```

```python
df_totals_max_year = df_rec_year[df_rec_year.recipient_parent.isin(["THE GEO GROUP INC", "CORRECTIONS CORPORATION OF AMERICA", "GEO GROUP INC., THE", "M V M  INC.", "CSI AVIATION SERVICES  INC.", "AHTNA  INCORPORATED", "CLASSIC AIR CHARTER INC.", "STG INTERNATIONAL  INC.", "ASSET PROTECTION & SECURITY SERVICES  L.P.", "SCIENCE APPLICATIONS INTERNATIONAL CORPORATION"])]
```

```python
df_totals_max_year.head(127)
```

```python
df_totals_by_year = df_totals_max_year.pivot_table(index="recipient_parent", columns="award_first_year", values="amount").reset_index().fillna(0)
df_totals_by_year
```

```python
df_totals_by_year.info()
```

```python
df_totals_by_year.to_csv("top_ten_totals.csv", index=False)
```
